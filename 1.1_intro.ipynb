```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename, "r") as f:
        data= json.load(f)
    return data
```

```python
#invoke
data= load_data("data.json")
```

```python
#function to diaplay users and their connections goal is to make the jason data as read able as possible
def display_users(data):
    print("Users and their connections \n")
    for user in data['users']:
        print(f"{user['name']} is friends with: {user['friends']} and liked pages: {user['liked_pages']}")
    print("\nPages Information \n")
    for page in data['pages']:
        print(f"{page['id']} : {page['name']}")

#call
display_users(data)
```
