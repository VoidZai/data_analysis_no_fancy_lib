```python
import json

#func to clean data
def clean_data(data):
    #1 remove users with missing names using list comprehention
    data["users"]= [user for user in data["users"] if user["name"].strip()]
    #2 remove duplicate friends by converting into set for unique values then converting it to list
    for user in data["users"]:
        user["friends"]= list(set(user["friends"]))
    #3 remove inactive users using list comprehention
    data["users"]= [user for user in data["users"] if user["friends"] or user["liked_pages"]]
    #4 remove duplicate pages using dictionary and set methods
    #logic here is to have uniquw ids but the value to be updated ex here 104 has 2 values ie 104: web hub and 104: web dev the logic is to have only one entry that is 104: web hub then to update the valyue from web hub to web dev
    unique_page= {}
    for page in data["pages"]:
       unique_page[page["id"]]= page
       data["pages"]= list(unique_page.values())
    return data

#load the data one liner no function
data= json.load(open("data2.json"))
#invoke
data= clean_data(data)
#convert the python object into json string and save it in a new file
json.dump(data, open("cleaned_data2.json", "w"), indent= 4) 
print("data file sucessfully cleaned and saved in cleaned_data2.json")
```
